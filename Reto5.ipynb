## Ejercicio 1: Funciones


De nuestro código que evaluaba cual era una el número mayor, convertirla a función.

Entrada: num1, num2

Salida: 

1. El num1 es mayor que num2
2. El num2 es mayor que num2
3. Ambos son iguales

````python
def mayor(num1, num2):
    if num1>num2:
        resultado=(f" {num1} es mayor que {num2}")
    elif num2>num1:
        resultado=(f" {num2} es mayor que {num1}")
    else:
        resultado=("ambos numeros son iguales")
    return resultado
````

**Ejemplo de impresión**

En el segundo `9` el piloto llego a los `223` km/h




````python
def veloz(lista):
    resultado=[]
    s=0
    for e in lista:
        s=s+1
        if e>= 210:
            resultado.append(s)
            print(f"en el segundo {s} la velocidad es {e}")
    return resultado

````

## Extra points

Hacer una función que nos imprima cuantos números pares e impares tiene una lista.

Para evaluar un par o impar puedes utilizar el

```Python
if numero%2 = 0:
    print("numero es par")
else:
    print("numero es impar")
```

|Entrada | Salida |
| - | - | 
| [2,3,6,8,9] | Tiene `3` pares y `2` inpares |

````python
def count_parOimpar(lista):
    par=0
    impar=0
    for e in lista:
        if e%2 == 0:
            #print("numero es par")
            par+=1
        else:
            #print("numero es impar")
            impar+=1
    
    return f"tiene {par} pares y {impar} impares "


lista=(2,3,6,8,9)
count_parOimpar(lista)
````

````python
# kata 4, recibir numero como str y devolver el mas grande y el mas pequeño como int

def high_and_low(numbers):
    lista=numbers.split(" ")
    enteros=[]
    for e in lista:
        enteros.append(int(e))

    return f"{max(enteros)} {min(enteros)} "

numeros=("1 32 -4 -56 4")

high_and_low(numeros)
````

````python
# kata 5, de una cadena contar las vocales (a, e, i, o, u)

def get_count(sentence):
    vocales=[]
    for e in sentence:
        if e=='a':
            vocales.append(e)
        elif e=='e':
            vocales.append(e)
        elif e=='i':
            vocales.append(e)
        elif e=='o':
            vocales.append(e)
        elif e=='u':
            vocales.append(e)
    return len(vocales)

cadena="hear me now u binary"

get_count(cadena)
````
