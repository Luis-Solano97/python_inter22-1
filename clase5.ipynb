## PROYECTO


```python
def generador_contraseña(longitud):
    letras="QWERTYUIOPASDFGHJKLÑZXCVBNMqwertyuiopasdfghjklñzxcvbnm"
    contra=""
    for e in range(longitud):
        contra+= random.choice(letras)
    return contra

longitud=int(input())
print(longitud)
generador_contraseña(longitud)
```

## list comprehension

```python
# listas permiten meter un for dentro
[i for i in range(10)]
# primero hace lo de la derecha, es decir el for
# despues lo de la izquierda
```

```python
[i/2 for i in range(10)]
```

```python
# se pueden meter condicionales
[i**2 for i in range(10) if i**2>25]
# de nuevo primero el for, luego sigue con el if y 
# si cumple la condicion ya ejecuta lo de la izquierda del for
```

```python
[letra for letra in 'auqiaosjiqlso' if letra in 'aeiou']
```
